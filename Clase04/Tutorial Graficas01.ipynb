### Propiedades de las grficas


<lu>
    <li>linewidth
    <li>alpha
    <li>color o c
    <li>dashes
    <li>linestyle o ls
    <li>marker
    <li>markeredgewidth, markeredgecolor
    <li>markerfacecolor, markersize
</lu>

```python
x=np.arange(0,5,0.1)        ##Genera valores de x entre 0 y 5 a intervalos de 0.1
plt.plot(x,np.exp(-x),linewidth=1,color='orange',marker='1')   ##linewidth ancho de linea
plt.plot(x,np.exp(-x*2),linewidth=2,dashes=[1,1,2,1])  ##dashes intermitente:  linea, espacio,linea, espacio
plt.plot(x,np.exp(-x*0.5),linewidth=4,alpha=0.5)  ##alpha de 0.0 a 1.0 transparencia
plt.text(1.0,0.5,"linewidth=4,alpha=0.5")    
plt.text(0.8,0.3,"linewidth=1")
plt.text(0.7,0.1,"linewidth=2")
plt.show()
```

```python
## markeredgecolor:color del borde del marcador
plt.plot([1,2,3],[2,4,9],marker='.',markersize=25,markeredgecolor='black')
## markeredgewidth: ancho del borde del marcador
plt.plot([1,2,3],[8,6,2],marker='.',markersize=25,markeredgecolor='yellow',markeredgewidth=4)
plt.show()
```

```python
## markerfacecolor: color del marcador
plt.plot([1,2,3],[2,4,9],marker='.',markerfacecolor='green',markersize=20.0)
plt.show()
```

```python
## markevery: coloca un marcador cada intervalo indicado 
## grid: coloca un enrrejado en la grfica
plt.plot([1,2,3,4,5,6,7,8,9],[2,4,9,7,2,8,3,1,5],marker='.',markerfacecolor='orange',markersize=15.0,markevery=2)
plt.grid(True)
plt.show()
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=True, facecolor='b', alpha=0.5,edgecolor='k')


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
